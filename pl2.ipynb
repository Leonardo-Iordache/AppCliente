```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.neural_network import MLPRegressor
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from numpy.ma.core import count
data = load_breast_cancer()

df = pd.DataFrame(data.data, columns=data.feature_names)
#0 para maligno, 1 para benigno
df['target'] = data.target

print(data.target)
```

```python
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)

mlp = MLPRegressor(solver='lbfgs',max_iter=10000, hidden_layer_sizes=(50, 50)) 

mlp.fit(X_train, y_train)
```

```python
y_pred_not_round = mlp.predict(X_train)
```

```python
y_pred2 = mlp.predict(X_test)

df_temp = pd.DataFrame({'Actual': y_train, 'Predicted': y_pred_not_round})


df_dif = pd.DataFrame({'Difference':df_temp['Actual'] - df_temp['Predicted']})
df_temp
```

```python
df_temp.plot(kind='line', figsize=(50, 25))
plt.show()
```
